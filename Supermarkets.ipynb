```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

area_size = input('Input area size:')
stock_count = input('Input stock count:')
daily_customer_count = input('Input daily customer count:')

data = pd.read_csv('supermarkets.csv')
data_input = data.drop(columns=['ID', 'Revenue'])
data_output = data['Revenue']

model = DecisionTreeClassifier()
model.fit(data_input.values, data_output.values)
predicted_revenue = model.predict([[area_size, stock_count, daily_customer_count]])
print('Predicted revenue: $%s' % (predicted_revenue[0]))
```
