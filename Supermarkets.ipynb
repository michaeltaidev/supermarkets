```python
# Revenue Prediction

import pandas as pd
from sklearn.tree import DecisionTreeRegressor

area_size = input('Input area size:')
stock_count = input('Input stock count:')
daily_customer_count = input('Input daily customer count:')

data = pd.read_csv('supermarkets.csv')
specs = data.drop(columns=['ID', 'Revenue'])
revenue = data['Revenue']

model = DecisionTreeRegressor()
model.fit(specs.values, revenue.values)
predicted_revenue = model.predict([[area_size, stock_count, daily_customer_count]])
print('Predicted revenue: $%s' % (predicted_revenue[0]))
```

```python
# Accuracy Analysis

import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split

data = pd.read_csv('supermarkets.csv')
specs = data.drop(columns=['ID', 'Revenue'])
revenue = data['Revenue']

train_specs, test_specs, train_revenue, test_revenue = train_test_split(specs, revenue, test_size=0.2)

model = DecisionTreeRegressor()
model.fit(train_specs, train_revenue)
accuracy_score = model.score(test_specs, test_revenue)
print('Accuracy Score: %s' % (accuracy_score))
```
